```python

bikes = {}

def add_bike(bike_id):
    if bike_id not in bikes:
        bikes[bike_id] = {"is_available": True}
        print(f"Bike with ID {bike_id} added to the system.")
    else:
        print(f"Bike with ID {bike_id} already exists in the system.")


def rent_bike(bike_id):
    if bike_id in bikes:
        if bikes[bike_id]["is_available"]:
            bikes[bike_id]["is_available"] = False
            print(f"Bike with ID {bike_id} rented successfully.")
        else:
            print(f"Bike with ID {bike_id} is not available for rent.")
    else:
        print(f"Bike with ID {bike_id} does not exist in the system.")

def return_bike(bike_id):
    if bike_id in bikes:
        if not bikes[bike_id]["is_available"]:
            bikes[bike_id]["is_available"] = True
            print(f"Bike with ID {bike_id} returned successfully.")
        else:
            print(f"Bike with ID {bike_id} is already available.")
    else:
        print(f"Bike with ID {bike_id} does not exist in the system.")

def display_menu():
    print("\n===== Bike Rental System Menu =====")
    print("1. Add a bike")
    print("2. Rent a bike")
    print("3. Return a bike")
    print("4. Exit")

def main():
    while True:
        display_menu()
        choice = input("Enter your choice: ")

        if choice == '1':
            bike_id = input("Enter the bike ID: ")
            add_bike(bike_id)
        elif choice == '2':
            bike_id = input("Enter the bike ID you want to rent: ")
            rent_bike(bike_id)
        elif choice == '3':
            bike_id = input("Enter the bike ID you want to return: ")
            return_bike(bike_id)
        elif choice == '4':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

main()
103
```
